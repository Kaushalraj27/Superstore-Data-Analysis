```python
# Superstore_EDA.py

"""
üìä Superstore Data Analysis ‚Äì Internship Project @ The Sparks Foundation

Performed during the Data Science & Business Analytics internship,
this script conducts an Exploratory Data Analysis (EDA) on the Sample Superstore dataset.
It identifies trends in sales, profits, and discount impacts across regions and categories.
"""

# --- üì¶ Import Required Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- üìÇ Load Dataset ---
file_path = r"E:\Data Science\All internship\Spark Foundation internship\Task-3 SampleSuperstore.csv"
df = pd.read_csv(file_path)

# --- üîç Initial Data Exploration ---
print("üìå Dataset Info:")
print(df.info())
print("\nüìå First 5 Rows:")
print(df.head())

# --- üßº Data Cleaning ---
# Drop unnecessary columns
df.drop(['Country', 'Postal Code'], axis=1, inplace=True)

# Check for missing values
print("\n‚ùì Missing Values:")
print(df.isnull().sum())

# Check for duplicates
df.drop_duplicates(inplace=True)

# --- üìä Exploratory Data Analysis ---

# 1. Sales vs Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales', y='Profit', hue='Category', data=df)
plt.title('Sales vs Profit by Category')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)
plt.show()

# 2. Top 10 States by Sales
top_states = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)
top_states.plot(kind='barh', color='skyblue')
plt.title('Top 10 States by Sales')
plt.xlabel('Sales')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# 3. Region-wise Profit Distribution
region_profit = df.groupby('Region')['Profit'].sum().sort_values()
region_profit.plot(kind='bar', color='orange')
plt.title('Profit by Region')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Segment-wise Sales and Profit
segment_analysis = df.groupby('Segment')[['Sales', 'Profit']].sum()
segment_analysis.plot(kind='bar', figsize=(8, 5))
plt.title('Sales and Profit by Segment')
plt.ylabel('Amount')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 5. Discount vs Profit
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Discount', y='Profit', marker='o')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True)
plt.show()

# --- ‚úÖ Summary ---
print("\n‚úÖ Key Observations:")
print("- Western region is the top contributor to profit.")
print("- Technology is the most profitable category.")
print("- High discounts often lead to losses.")
print("- Consumer segment dominates in revenue.")
print("- States like California and New York outperform others.")

# --- üì¨ Contact Info ---
print("\nüì´ For any queries, reach out to Kaushal Raj ‚Äì kaushal.raj.tech22@gsv.ac.in")

# --- END OF SCRIPT ---
```
